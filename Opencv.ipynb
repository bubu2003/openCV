```python
# step 1 import required library
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow # for image display in colab

# step2 upload image
from google.colab import files
uploaded = files.upload()
```

```python
#3 read the img
filename=list(uploaded.keys())[0]
image=cv2.imread(filename) #open cv reads the bgr format
cv2_imshow(image) #show the orignal img
```

```python
#4 convert into gray
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
#5 resize img 300x300
resized=cv2.resize(gray,(300,300))
cv2_imshow(resized)
```

```python

a=[image,gray]
for i in a:
  resized=cv2.resize(i,(300,300))
  cv2_imshow(resized)
```

```python
# draw reactangle and text
annotated=image.copy()
cv2.rectangle(annotated,(30,0),(200,200),(0,255,0),5)
cv2.putText(annotated,'Google', (60,35),cv2.FONT_HERSHEY_SIMPLEX,1,(127,255),5)
cv2_imshow(annotated)
```

```python
#7 save img

cv2.imwrite('output_image.png',annotated)
print('image saved as output_image.png')
```

project 2 data augumentation


```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img,img_to_array
```

```python
#3 UPLOAD IMG
uploaded=files.upload()
filename=list(uploaded.keys())[0]
```

```python
#4 load and prepare img
img=load_img(filename,target_size=(224,224))# resize img
img_array=img_to_array(img)
img_array=np.expand_dims(img_array,axis=0) # convert to batch 1
```

```python
# define augumentation generator
datagen=ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    brightness_range=(0.5,1.5)
)

#generate augumentation img
i=0
for batch in datagen.flow(img_array,batch_size=1):
  plt.figure(figsize=(2,2))
  plt.imshow(batch[0].astype('uint8'))
  i+=1
  if i>7:
    break
```
